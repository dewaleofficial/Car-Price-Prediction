```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('clean_car_df1.csv')
```

```python
df.head()
```

```python
np.sum(df.isnull())
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
for i in df['make']:
    try:
        result = i.lstrip()
    except:
        i
    print(result)
```

```python
df_1 = df.copy()

df_1["car_type"] = df_1["make"].apply(lambda x: x.lstrip())
```

```python
df_1["car_make"] = df_1["car_type"].apply(lambda x: x.split(' '))
```

```python
df_1["car_make"].head()
```

```python
df_1["car_name"] = df_1["car_make"].apply(lambda x: x[0])
```

```python
df_1["model"] = df_1["car_make"].apply(lambda x: x[1])
```

```python
df_1.head()
```

```python
df.drop('make', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['car_name'] = df_1['car_name']
df['model'] = df_1['model']
```

```python
df.head()
```

```python
cars_data = df[['car_name', 'model', 'mileage(Km)', 'year', 'condition', 'price(₦)']]
```

```python
cars_data.head()
```

```python
cars_data.to_csv('cars_data.csv', index=False)
```
